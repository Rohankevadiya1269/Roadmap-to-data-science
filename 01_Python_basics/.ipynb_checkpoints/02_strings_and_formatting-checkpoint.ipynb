# Strings and Formatting

This notebook covers how text data (strings) are handled in Python.
Strings are critical in data science because most real-world data conatins names, categories, labels and free-text fields.

## What is a string?
A string is a sequence of characters used to represent text. In python, Strings are enclosed in quotes.

In real-world data science, strings appear in:
- Names (customer, products)
- Categories(department, region)
- Labels(Yes/No, Active/Inactive)
- Free-text fields(reviews, comments)

```python
# creating string variables 
first_name = "Rohan"
last_name = "Kevadiya"
program = "Ms. in Data Science"
# Strings are immutable i.e cannot be changed in place 
print(first_name)
print(last_name)
print(program)
```

```python
# String concatenation

#combining strings using concatenation
full_name = first_name  + " " + last_name
print(full_name)
# The " " in the concatenation part is for the space in between first and last name
```

```python
# using f-strings for formatting
message = f"My full name is {full_name} and I study {program}."
print(message)
```

```python
# Accessing individual characters in string using Indexing
# Indexing starts at 0. which means instead of 1, 0 is the first index
# word[0] is first char and word[-1] is last char
word = "Python"
print(word[0])
print(word[2])
print(word[-1])
```

```python
#Extracting substrings using slicing
# variable [start:end]
print(word[0:3])
print(word[2:6])
```

```python
# A few examples of negative indexing
newWord = "DataScience"
print(newWord[-2]) # prints second last character
print(newWord[-7:-1])

# Using one positive and one negative value in indexing
print(newWord[4:-4])

# Example of user id
citycode = user_id[4:-5]
print(citycode)

# removing file extension
file = "final_report_v2.pdf"
base_name = file[:-4]

# mixing indexing with string method .find()
user_mail = "miscellenoususer123@gmail.com"
print(user_mail[:user_mail.find("@")])

```

```python
#Common string methods
text = "data Science is powerful"
print(text.upper())
print(text.lower())
print(text.strip())  # Strip() removes unwanted spaces
```

```python
# More String Methods
# Replace
text = 'data-science'
clean_text = text.replace("-", " ")
print("\n", clean_text)

record = "USA,Chicago,2025"
parts = record.split(",") # Use case - parsing CSV-like data, splitting IDs, Feature extraction
print(parts)
print(parts[1])

words = ["Data", "Science", "Roadmap"]
sentence = " ".join(words)
print("\n", sentence)

file_name = "final_Exam_report.pdf"
print(file_name.endswith(".pdf"))
print(file_name.startswith("final_Exam"))

# isdigit(), isalpha(), isalnum()
value = "23453211"
print(value.isdigit())

name = "Rohan"
print(name.isalpha())

fields = "data science data analysis"
print(fields.count("data"))

email = "rohan@gmail.com"
print(email.find("@"))


print(text.capitalize())
print(text.title())




```

```python
#Taking User string input
user_name = input("Enter your name: ")
print(f"Welcome, {user_name}!")
```

```python
# Example on how parsing and validatinng connects in real world
user_id = "USR-CHI-8921"
if user_id.startswith("USR-"):
    city_code = user_id[4:7]
    print(city_code)
```

### Important terminologies :
Parsing - Breaking string into meaningful parts to use them later.
 Validating - Checking whether a string follows expected patterns. These paterns are already defined.
 In slicing the start is INCLUSIVE(meaning the first character or the start index is included). And the end is EXCLUSIVE(meaning the end index is not included, in fact python stops before the end index).
