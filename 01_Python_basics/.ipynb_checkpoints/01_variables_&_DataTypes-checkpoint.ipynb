## What is a Variable?

A variable is a named container used to store data in memory.
Instead of repeating values, we store them once and reuse them.

In data science, variables are used to store:
- Raw data values
- Intermediate results
- Features used in models


## What is a Data Type?

A data type defines the kind of value a variable holds.
Using correct data types is critical for calculations and logical operations.

Common data types in Python:
- int   → whole numbers
- float → decimal numbers
- str   → text values
- bool  → True or False values


```python
#Storing Basic information in variables
Name = "Rohan Kevadiya"
program = "MS. Data Science"
credits_completed = 12
current_gpa = 3.4
actively_applying = False

#printing values to the console
print(Name)
print(program)
print(credits_completed)
print(current_gpa)
print(actively_applying)
```

#Inspecting the Data Types
print(type(Name))
print(type(program))
print(type(credits_completed))
print(type(current_gpa))
print(type(actively_applying))


## What this teaches you

Python automatically infers data types. 
Data scientists frequently check types to avoid silent bugs. 
This is basic data validation, not “just printing”
